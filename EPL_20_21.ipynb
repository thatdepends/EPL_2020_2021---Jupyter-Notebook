```python
file_path = 'C://Users//Uporabnik//jupyter//epl_20_21.csv'
epl_df = pd.read_csv(file_path)
epl_df.head() 
#print(os.path.exists(file_path))
```

```python
#Create 2 new columns

epl_df['MinsPerMatch'] = (epl_df['Mins'] / epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch'] = (epl_df['Goals'] / epl_df['Matches']).astype(float)
print(epl_df.head())      
```

```python
#Total goals

Total_Goals = epl_df['Goals'].sum()
print(Total_Goals)
```

```python
#Penalty Goals

Toatal_PenaltyGoals = epl_df['Penalty_Goals'].sum()
print(Toatal_PenaltyGoals)
```

```python
#Penalty Attempts

Total_PenaltyAttempts = epl_df['Penalty_Attempted'].sum()
print(Total_PenaltyAttempts)
```

```python
# Pie Chart for penalty missed vs scored

pen_not_scored = epl_df['Penalty_Attempted'].sum() - Toatal_PenaltyGoals
data = [pen_not_scored, Toatal_PenaltyGoals]

# Create labels
labels = ['Penalty Missed', 'Penalty Scored']

# Set color palette
color = sns.color_palette('Set3')

# Plotting the pie chart
plt.figure(figsize=(13, 6))
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
plt.title('Penalty Missed vs Scored')
plt.show()
```

```python
# Count the number of players per club and get the top 5 clubs
top_clubs = epl_df['Club'].value_counts().nlargest(5)

# Create a bar plot
plt.figure(figsize=(10, 6))
top_clubs.plot(kind='bar', color=sns.color_palette("viridis"))
plt.title('Top 5 Clubs with Maximum Players')
plt.xlabel('Club')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()
```

```python
# Define age bins and labels
bins = [0, 20, 25, 30, float('inf')]  # Using float('inf') for the upper limit of 'above 30'
labels = ['Under 20', '20-25', '25-30', 'Above 30']

# Create a new column for age groups
epl_df['Age_Group'] = pd.cut(epl_df['Age'], bins=bins, labels=labels, right=False)

# Now you can easily group or filter players based on their age group.
# For example:
under20 = epl_df[epl_df['Age_Group'] == 'Under 20']
age20_25 = epl_df[epl_df['Age_Group'] == '20-25']
age25_30 = epl_df[epl_df['Age_Group'] == '25-30']
above30 = epl_df[epl_df['Age_Group'] == 'Above 30']
```

```python
x = np.array([
    under20['Name'].count(), 
    age20_25['Name'].count(), 
    age25_30['Name'].count(), 
    above30['Name'].count()
])

mylabels = ["<= 20", "20 - 25", "25 - 30", "Above 30"]

# Create the pie chart
plt.figure(figsize=(10, 6))
plt.pie(x, labels=mylabels, autopct="%.1f%%", startangle=140, colors=sns.color_palette('Set3'))

# Add a title
plt.title('Total Players by Age Group', fontsize=20)
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle
plt.show()
```

```python
# Filter players under 20
players_under_20 = epl_df[epl_df['Age'] < 20]

# Count the number of players per club
club_counts = players_under_20['Club'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
club_counts.plot(kind='bar', color=sns.color_palette('cubehelix'))

# Adding title and labels
plt.title('Number of Players Under 20 in Each Club', fontsize=20)
plt.xlabel('Club', fontsize=14)
plt.ylabel('Number of Players', fontsize=14)

# Show the plot
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.tight_layout()  # Adjust layout to make space for labels
plt.show()
```

```python
# avg ages of players in each club

# Set the figure size
plt.figure(figsize=(12, 6))

# Create the box plot
sns.boxplot(x='Club', y='Age', data=epl_df)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=90)

# Add titles and labels
plt.title('Distribution of Players\' Ages by Club', fontsize=16)
plt.xlabel('Club', fontsize=14)
plt.ylabel('Age', fontsize=14)

# Show the plot
plt.tight_layout()  # Adjust layout to accommodate labels
plt.show()
```

```python
# Calculate total assists from each club
Assists_by_clubs = epl_df.groupby('Club', as_index=False)['Assists'].sum()

# Set the theme for seaborn
sns.set_theme(style="whitegrid", color_codes=True)

# Set the figure size
plt.figure(figsize=(20, 8))

# Create the bar plot
ax = sns.barplot(x='Club', y='Assists', data=Assists_by_clubs.sort_values(by='Assists', ascending=False), palette='Set2')

# Set the x and y labels
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Total Assists', fontsize=20)

# Rotate x-axis labels for better readability
plt.xticks(rotation=75)

# Set the title of the plot
plt.title('Total Assists by Club', fontsize=20)

# Show the plot
plt.show()
```
